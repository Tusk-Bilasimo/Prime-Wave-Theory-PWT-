```python
# Balmer Series Simulation with Prime Signatures
# This simulates hydrogen Balmer series wavelengths and factors them into primes,
# tying to PWT's Harmonic Cascade.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Prime factorization function
def prime_signature(n):
    return sp.factorint(n)

# Hydrogen spectral lines (Balmer series, n=3 to 9)
def balmer_wavelength(n):
    R = 1.097e7  # Rydberg constant (1/m)
    return 1 / (R * (1/4 - 1/n**2)) * 1e9  # in nm

n_values = np.arange(3, 10)
wavelengths = [round(balmer_wavelength(n)) for n in n_values]  # Round to integer nm for factorization

# Factorize wavelengths
factors = [prime_signature(wl) for wl in wavelengths]

# Plot
plt.plot(n_values, wavelengths, marker='o')
plt.title('Balmer Series Wavelengths')
plt.xlabel('Principal Quantum Number n')
plt.ylabel('Wavelength (nm)')
plt.grid(True)
plt.savefig('balmer_plot.png')  # For blog embed
plt.show()

# Output for blog
print("Wavelengths (nm):", wavelengths)
print("Prime Signatures:", factors)
```
